```python
total_victims_state = MS.groupby('State').sum()
print ('Total Victims = ', total_victims_state['Total'].sum())
total_victims_state.sort_values(by = 'Total', ascending = False)
```

```python
total_victims_state.sort_values(by = 'Total', ascending = False).tail()
```

```python
plt.subplots(figsize = (15, 6))
shooting = total_victims_state['Total'].sort_values(ascending = False)
ax = shooting.plot.bar()
ax.set_xlabel('State')
ax.set_ylabel('Total Number = Killed + Injured')
ax.set_title('Total Reported Cases by State')
plt.show()
print(shooting)
```

```python
import squarify

y = MS['Total'].value_counts().head(10)
plt.rcParams['figure.figsize'] = (10,10)
plt.style.use('fivethirtyeight')
color = plt.cm.magma(np.linspace(0,1,15))
squarify.plot(sizes = y.values, label = y.index, alpha=.8, color=color, pad = True)
plt.title('Tree Map for Crime', fontsize = 20)
plt.axis('off')
plt.show()
```

```python
plt.style.use('seaborn')
MS['Incident Date'].value_counts().head(3).plot.pie(figsize = (15,8))
plt.title('Top 3 Dates With The Highest Mass Shootings', fontsize = 20)
plt.xticks(rotation = 90)
plt.show()
```

```python
MS['Incident Date'] = pd.to_datetime(MS['Incident Date'])
MS['Month'] = MS['Incident Date'].dt.month
plt.style.use('fivethirtyeight')
plt.rcParams['figure.figsize'] = (15, 8)
sns.countplot(MS['Month'], palette = 'autumn')
plt.title('Highest Mass Shootings by Month')
```
